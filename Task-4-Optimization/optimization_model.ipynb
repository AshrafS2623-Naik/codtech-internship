```python
from pulp import *

model = LpProblem("Profit_Maximization",LpMaximize)

A=LpVariable("A",lowBound=0)
B=LpVariable("B",lowBound=0)

model+=20*A+30*B

model+=2*A+1*B<=100
model+=1*A+2*B<=80

model.solve()

print("A =",A.varValue)
print("B =",B.varValue)
print("Profit =",value(model.objective))
```
